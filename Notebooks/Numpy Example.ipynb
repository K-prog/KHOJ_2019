```python
import matplotlib.pyplot as plt
import numpy as np
N = 8
y = np.zeros(N)
```

```python
x1 = np.linspace(0, 10, N, endpoint=True)
print(x1)
```

```python
x2 = np.linspace(0, 10, N, endpoint=False)
print(x2)
```

```python
plt.plot(x1, y, 'o')
# plt.plot(x2, y + 0.5, 'o')
plt.ylim([-0.5, 1])
plt.show()
```

```python
plt.plot(x2, y + 0.5, 'o')
plt.ylim([-0.5, 1])
plt.show()
```
